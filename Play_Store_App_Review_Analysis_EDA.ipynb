```python
df['Size'].unique()
```

```python
#Function to convert KB to MB
def Kb_to_Mb(val):
  if val == 'Varies with device':
    val = 'Varies with device'
  else:
    if val[-1] == 'k':
      val = round(float(val[:-1])/1024,3);        # 1MB = 1024KB
    else:
      val = float(val[:-1])
  return val
```

```python
df['Size'] = df['Size'].apply(lambda x: Kb_to_Mb(x))
df.head()
```
